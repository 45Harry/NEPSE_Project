```python
import requests
from bs4 import BeautifulSoup


# Fetch the webpage content
url = "https://nepsealpha.com/"
response = requests.get(url)

# Parse the content with BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")
print("BeautifulSoup object created successfully.")
```

```python
# Find the first table on the page
first_table = soup.find("a")
print(first_table)  # Prettified HTML of the first table

```

```python
import cloudscraper
import time

def fetch_data(url):
    scraper = cloudscraper.create_scraper()
    response = scraper.get(url)
    #response = response.encoding('utf-8')
    
    if "Cloudflare" in response.text:
        print("Blocked by Cloudflare! Retrying with delay...")
        time.sleep(5)
        return fetch_data(url)  # Recursive retry
    
    return response
```

```python
json_data = fetch_data("https://www.nepsealpha.com/")
```
