```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://merolagani.com/")
response.encoding = "utf-8"  # Ensure proper encoding for Nepali characters
soup = BeautifulSoup(response.text, "html.parser")
```

```python
headlines = [element.text for element in soup.select("#ctl00_ContentPlaceHolder1_tblRecentNews a")]
headlines
```

```python
import json

# Ensure proper encoding for Nepali characters
with open("news.json", "w", encoding="utf-8") as file:
    json.dump(headlines, file, indent=4, ensure_ascii=False)
```

```python
link = "https://merolagani.com/NewsDetail.aspx?newsID=116311"
response = requests.get(link)
response.encoding = "utf-8"  # Ensure proper encoding for Nepali characters

soup = BeautifulSoup(response.text, "html.parser")
body = soup.find_All("div", class_=["media-news", "media-news-lg", "alt"])
body
```

```python
# Extract text from all elements with class 'media-content'
media_contents = soup.select('.media-content')
media_text = [content.get_text(strip=True) for content in media_contents]
'\n\n'.join(media_text)
```
